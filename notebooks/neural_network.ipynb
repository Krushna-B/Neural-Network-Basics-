# One Neuron

```python
import numpy as np 

def relu(x):
    return np.maximum(x,0)

def neuron(w,x,b):
    z =  np.dot(w,x) + b
    return relu(z)

x = np.array([10.1,0,-3,9,0.1]) # d x 1
w = np.array([0.1,0.5,0.3,0.8,0.9]) # d x 1
b = 0.1



print(neuron(w,x,b))



```

# One Layer

```python
# Layer of m neuron's, is weights of m x d, each neruon as d dimentional weights
# b is dimensions m, each neuron has 1 bias

def create_layer(m,d,X):
    W = np.random.randn(m,d) * 0.1 # Creat m x d dimensions wieghts for layer
    b = np.random.randn(m) * 0.1

    Z = W @ X + b

    print(f"Output before activation is {Z}")
    print(Z.shape)
    return(relu(Z))


X = np.array([1,2,3,4,5]) #5 dimension input

print(create_layer(5,5,X))




```

# Neural Network 

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

X = np.array([ [0.1,0.5,0.7],  #Input of shape 2 x 3
              [0.8,0.9,0.10]
])
y = np.array([1,0]) #Actual values 
np.random.seed(10)

W1 = np.random.randn(10,3) # Shape 10 x 3 
b1 = np.random.rand(10,1)

W2 = np.random.randn(1,10) # Shape 4 x 10
b2 = np.random.rand(1,1)


z1 = W1 @ X.T + b1
a1 = relu(z1)

z2 = W2 @ a1 + b2
y_hat = sigmoid(z2)
print(y_hat)






```

# Loss Function (Binary Cross Entropy Loss Function)

```python
def binary_cross_entropy(y,y_hat):
    eps = 1e-8
    return -np.mean(y * np.log(y_hat + eps) + (1-y)*np.log(1-y_hat + eps))
```
